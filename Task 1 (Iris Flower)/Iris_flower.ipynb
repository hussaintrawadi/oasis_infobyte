Visualising Data

```python
plt.figure(figsize=(20, 8))
sns.countplot(data=df, x='sepal_length', hue='species', palette='inferno')
plt.show()
```

In conclusion we can say density of Sentosa is more on lower sepal length and density of virginica is more on higher sepal length

```python
plt.figure(figsize=(20,8))
sns.countplot(data=df, x='sepal_width', hue='species',palette = 'Reds')
```

In conclusion we can say that density of Versicolor is more on lower sepal width and density of sentosa is more on higher sepal width

```python
plt.figure(figsize=(20,8))
sns.countplot(data=df, x='petal_length', hue='species', palette= 'magma')
```

In conclusion we can say that density of Sentosa is more on lower petal length and density of virginica is more on higher petal length

```python
plt.figure(figsize = (6,6))
sns.scatterplot(x = 'sepal_length', y = 'sepal_width' , data = df , hue = 'species', palette = 'inferno',s= 50)
```

In conclusion we can say that Sentosa have higher sepal length but lower sepal width, Versicolor is moderate in comparasion to sepal height and sepal width & Virgincia has higher sepal width but lower sepal length

```python
plt.figure(figsize=(1,1))
sns.pairplot(df)
```

Building machine learning model (Linear Regression)

```python
x = df.drop(['species'] , axis = 1)
y = df['species']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
lr.score(x_train,y_train)
```

```python
predictions = lr.predict(x_test)
predictions
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

The accuracy is 97.7%
